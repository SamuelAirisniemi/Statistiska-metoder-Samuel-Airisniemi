```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng()
```

# A)

```python
data = {
    "Integers (0-100)": rng.integers(100, size = 100),
    "Uniform (0,1)": rng.uniform(size = 100),
    "Normal (0,1)": rng.normal(size = 100),
    "Binomial (n = 10, p = 0.3)": rng.binomial(n = 10, p = 0.3, size = 100),
    "Negative Binomial (n = 10, p = 0.3)": rng.negative_binomial(n = 10, p = 0.3, size = 100),
    "Gamma (shape = 1, scale = 0.9)": rng.gamma(shape = 1, scale = 0.9, size = 100),
    "Geometric (p = 0.31)": rng.geometric(p = 0.31, size = 100)
}

fig, axes = plt.subplots(3, 3, figsize = (14, 12))
axes = axes.flatten()

for ax, (title, values) in zip(axes, data.items()):
    ax.hist(values, bins = 15, color = "steelblue", edgecolor = "black")
    ax.set_title(title, fontsize = 12)
    ax.set_xlabel("")
    ax.set_ylabel("Count")

for i in range(len(data), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()
```

# B)

```python
data = {
    "Integers (0-100)": rng.integers(100, size = 100),
    "Uniform (0,1)": rng.uniform(size = 100),
    "Normal (0,1)": rng.normal(size = 100),
    "Binomial (n = 10, p = 0.3)": rng.binomial(n = 10, p = 0.3, size = 100),
    "Negative Binomial (n = 10, p = 0.3)": rng.negative_binomial(n = 10, p = 0.3, size = 100),
    "Gamma (shape = 1, scale = 0.9)": rng.gamma(shape = 1, scale = 0.9, size = 100),
    "Geometric (p = 0.31)": rng.geometric(p = 0.31, size = 100)
}

fig, axes = plt.subplots(3, 3, figsize = (14, 12))
axes = axes.flatten()

for ax, (title, values) in zip (axes, data.items()):
    sorted_vals = np.sort(values)
    cdf = np.arange(1, len(values) + 1) / len(values)
    ax.plot(sorted_vals, cdf, color = "steelblue", linewidth = 2)
    ax.set_title(title, fontsize = 12)
    ax.set_xlabel("Value")
    ax.set_ylabel("Cumulative Probability")
    ax.grid(True, linestyle = "--", alpha = 0.5)

for i in range(len(data), len(axes)):
    axes[i].remove()


plt.tight_layout()
plt.show()
```
